```python
from datetime import datetime
```

```python
def todatetime(s):
    if s != "0000-00-00 00:00:00":
        return datetime.fromisoformat(s)
```

```python
df["SessionStartTime"] = df["SessionStartTime"].apply(todatetime)
df["EventTime"] = df["EventTime"].apply(todatetime)
```

```python
df['SessionStartWeek'] = df['SessionStartTime'].dt.week
```

```python
week = df[['SessionStartWeek', 'ChargeInUSDAmount']].groupby('SessionStartWeek', as_index=False).agg(np.sum).sort_values(by='ChargeInUSDAmount')
```
